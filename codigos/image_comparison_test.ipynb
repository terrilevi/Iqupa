# Operaciones con imagenes

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_homero = cv2.imread("./homero1.png")
img_trash = cv2.imread("./trash1.png")

img_homero = cv2.cvtColor(img_homero, cv2.COLOR_BGR2RGB)
img_trash = cv2.cvtColor(img_trash, cv2.COLOR_BGR2RGB)

#dst = cv2.add(img_trash,img_homero)
#plt.imshow(img_trash)
#img_trash.shape, img_homero.shape

height1, width1 = img_homero.shape[0],img_homero.shape[1]
print(height1, width1)

height2, width2 = img_trash.shape[0],img_trash.shape[1]
print(height2, width2)
```

```python
min_height = min(height1, height2)
min_width = min(width1, width2)
print(min_height, min_width)
```

```python
# Crop images with minimum height and width
img11 = img_homero[0:min_height, 0:min_width]
img22 = img_trash[0:min_height, 0:min_width]
```

```python
plt.imshow(img11)
plt.show()
```

```python
plt.imshow(img22)
plt.show()
```

```python
# Creatte ROI by adding the cropped images
img_add = cv2.add(img11, img22)
# Show the added miminum images
plt.imshow(img_add)
plt.show()
```

```python
# Add minimum images with different weights
img_add = cv2.addWeighted(img11, 0.1, img22, 0.9, 0) 
plt.imshow(img_add)
plt.show()
```

```python
img_homero[0:min_height,0:min_width] = img_add
img_trash[0:min_height,0:min_width] = img_add

plt.imshow(img_homero)
plt.show()
```
